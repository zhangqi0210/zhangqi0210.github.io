```python
import numpy as np
import pandas as pd
```

```python
def loss_ratio(lst):
    df1 = pd.DataFrame(df_claim.groupby(lst).sum())
    df2 = pd.DataFrame(df_premium.groupby(lst).sum()['Total Premium'])
    dftemp = pd.DataFrame(df1['Claim Amount'])
    dftemp['Expense Amount'] = df1['Expense Amount']
    dftemp['Total Premium'] = pd.DataFrame(df2['Total Premium'])
    dftemp['Loss Ratio'] = dftemp['Claim Amount']/dftemp['Total Premium']
    dftemp['Combined Ratio'] = (dftemp['Claim Amount']+dftemp['Expense Amount'])/dftemp['Total Premium']
    dftemp = dftemp.reset_index()
    return dftemp
```

```python
def change_group(df):
    df.loc[df['group'] == 0, 'group'] = 'Under 18'
    df.loc[df['group'] == 1, 'group'] = '18-24'
    df.loc[df['group'] == 2, 'group'] = '25-34'
    df.loc[df['group'] == 3, 'group'] = '35-44'
    df.loc[df['group'] == 4, 'group'] = '45-54'
    df.loc[df['group'] == 5, 'group'] = '55-64'
    df.loc[df['group'] == 6, 'group'] = 'Above 65'
    return df
```

```python
df_premium = pd.read_csv('Premium.csv',low_memory=False, index_col = 0)
df_premium['group'] = 0
df_premium.loc[df_premium['age'] < 18, 'group'] = 0
df_premium.loc[(df_premium['age'] >= 18) & (df_premium['age'] <= 24), 'group'] = 1
df_premium.loc[(df_premium['age'] >= 25) & (df_premium['age'] <= 34), 'group'] = 2
df_premium.loc[(df_premium['age'] >= 35) & (df_premium['age'] <= 44), 'group'] = 3
df_premium.loc[(df_premium['age'] >= 45) & (df_premium['age'] <= 54), 'group'] = 4
df_premium.loc[(df_premium['age'] >= 55) & (df_premium['age'] <= 64), 'group'] = 5
df_premium.loc[df_premium['age'] >= 65, 'group'] = 6
df_claim = pd.read_csv('Claim.csv',low_memory=False, index_col = 0)
df_claim['group'] = 0
df_claim.loc[df_claim['age'] < 18, 'group'] = 0
df_claim.loc[(df_claim['age'] >= 18) & (df_claim['age'] <= 24), 'group'] = 1
df_claim.loc[(df_claim['age'] >= 25) & (df_claim['age'] <= 34), 'group'] = 2
df_claim.loc[(df_claim['age'] >= 35) & (df_claim['age'] <= 44), 'group'] = 3
df_claim.loc[(df_claim['age'] >= 45) & (df_claim['age'] <= 54), 'group'] = 4
df_claim.loc[(df_claim['age'] >= 55) & (df_claim['age'] <= 64), 'group'] = 5
df_claim.loc[df_claim['age'] >= 65, 'group'] = 6
```
